```python
# prompt: 위에 짠 코드 이해할 수 있게 설명을 적어줄래?

import random # 랜덤 숫자 생성을 위해 random 모듈을 import 합니다.

def 숫자_맞추기_게임(): # 숫자 맞추기 게임 함수를 정의합니다.
  숫자 = random.randint(1, 100) # 1부터 100 사이의 랜덤 숫자를 생성합니다.
  시도_횟수 = 0 # 사용자의 시도 횟수를 0으로 초기화합니다.

  print("1부터 100 사이의 숫자를 맞춰보세요!") # 게임 시작 메시지를 출력합니다.

  while True: # 무한 루프를 시작합니다.
    try: # 사용자 입력을 받는 부분에서 에러 처리를 시도합니다.
      추측 = int(input("숫자를 입력하세요: ")) # 사용자로부터 숫자를 입력받습니다.
    except ValueError: # 입력값이 숫자가 아닌 경우 에러 메시지를 출력합니다.
      print("잘못된 입력입니다. 숫자를 입력해주세요.")
      continue # 루프의 처음으로 돌아갑니다.

    시도_횟수 += 1 # 시도 횟수를 1 증가시킵니다.

    if 추측 < 숫자: # 사용자의 추측이 정답보다 작은 경우
      print("너무 낮습니다!") # "너무 낮습니다!" 메시지를 출력합니다.
    elif 추측 > 숫자: # 사용자의 추측이 정답보다 큰 경우
      print("너무 높습니다!") # "너무 높습니다!" 메시지를 출력합니다.
    else: # 사용자의 추측이 정답과 같은 경우
      print(f"축하합니다! {시도_횟수}번 만에 맞추셨습니다.") # 축하 메시지와 함께 시도 횟수를 출력합니다.
      break # 루프를 종료합니다.

숫자_맞추기_게임() # 숫자 맞추기 게임 함수를 호출하여 게임을 시작합니다.

# 이 코드는 1부터 100 사이의 숫자를 맞추는 간단한 게임입니다.
# 사용자는 숫자를 입력하고 프로그램은 사용자의 입력이 정답보다 높은지 낮은지 알려줍니다.
# 사용자가 정답을 맞추면 프로그램은 축하 메시지와 함께 시도 횟수를 출력합니다.
```
